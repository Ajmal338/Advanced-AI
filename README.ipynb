## 2. Data Preprocessing

### Import the dataset 

`x_train` is the dataset of 32x32 color images of objects that the model will be trained on.

`y_train` is the dataset of labels that correspond to `x_train`.

`x_test` is the dataset of 32x32 color images of objects that the model will be tested on.

`y_test` is the dataset of labels that correspond to `x_test`.

```python
cifar10 = tf.keras.datasets.cifar10 #import dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
y_train = y_train.flatten() #flatten data to convert to 1D image vectors
y_test = y_test.flatten()
```

### How many images are in each class

```python
classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10,7)) #plot all the classes
p = sns.countplot(y_train)
```

### Check for NaN values in our data

```python
np.isnan(x_train).any()
```

```python
np.isnan(x_test).any()
```

### Normalize and Reshape data

```python
input_shape = (32, 32, 3)

#reshaped t0 4D
#we want each value to be between 0.0 and 1.0. As all values originally fall under the 0.0-255.0 range, divide by 255.0.
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 3)
x_train = x_train / 255.0

x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 3)
x_test = x_test / 255.0
```

### Label encoding

```python
#convert to one-hot encodings, to include categorical dataset in our model
y_train = tf.one_hot(y_train.astype(np.int32), depth=10)
y_test = tf.one_hot(y_test.astype(np.int32), depth=10)
```

## 3. CNN

### 3.1 Define the model (1st)

#### Basic CNN structure is as follows: *Conv-Conv-Pool-Drop-Conv-Conv-Pool-Drop*

>The following cell defines `batch_size`, `num_classes`, and `epochs`.

```python
#Initialize variables
batch_size = 32
num_classes = 10
epochs = 50
```

```python
model_1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, 3, padding='same', input_shape=x_train.shape[1:], activation='relu'),
    tf.keras.layers.Conv2D(32, 3, activation='relu'), #relu is the rectifier, and it is used to find nonlinearity in the data.
    tf.keras.layers.MaxPooling2D(), #MaxPool2D is a downsampling filter.
    tf.keras.layers.Dropout(0.25),  #Regularization Layer with 25% dropout

    tf.keras.layers.Conv2D(64, 3, padding='same', activation='relu'),
    tf.keras.layers.Conv2D(64, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Dropout(0.25),

    tf.keras.layers.Flatten(),  #Converts tensors into 1D vectors
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(num_classes, activation='softmax'),
])

# Customizable Parameters
model_1.compile(optimizer=tf.keras.optimizers.legacy.RMSprop(learning_rate=0.0001, decay=1e-06),
            loss='categorical_crossentropy', metrics=['acc'])
```

### 3.2 Display the Architecture of the Model

```python
model_1.summary() #Display Model Architecture
```

### 3.3 Fit the training data

```python
#Fit the model
history_1 = model_1.fit(x_train, y_train, validation_data = (x_test, y_test), batch_size = batch_size, epochs = epochs)
```

## 4. Evaluate the Model, and Change the Parameters

### 4.1 Loss and Accuracy Curves

```python
#Plot Learning curves for the model
history_df_1 = pd.DataFrame(history_1.history)

fig, ax = plt.subplots()
history_df_1.plot(ax=ax)
ax.set(title='Model Learning Curves', xlabel='Epoch', ylabel='Accuracy/Loss')
plt.show()
```

### 4.2 Predict Results

```python
#Predict Results
test_loss, test_acc = model_1.evaluate(x_test, y_test)
```

>Our model runs with an accuracy of **78%** on our testing data

### 4.3 Confusion Matrix

```python
# Predict the values from the validation dataset
y_pred = model_1.predict(x_test)
# Convert predictions classes to one hot vectors
y_pred_classes = np.argmax(y_pred, axis = 1)
# Convert validation observations to one hot vectors
y_true = np.argmax(y_test, axis = 1)
# Compute the confusion matrix
confusion_matrix_1 = tf.math.confusion_matrix(y_true, y_pred_classes)
```

```python
#Plot results
plt.figure(figsize=(12,9))
a = sns.heatmap(confusion_matrix_1, annot = True, fmt = 'g')
a.set(xticklabels = classes, yticklabels = classes)
```

We see that our model classifies automobiles quite well, with **875** out of **1000** automobile images being classified correctly. In addition to, we see a relatively high confusion between cats and dogs.

## 5. CNN (2nd) 

### 5.1 Define the model (2nd)

#### Basic CNN structure is as follows: *Conv-Pool-Conv-Pool-Conv-Pool-Conv-Pool* with doubling the number of filters every stage

>The following cell defines `batch_size`, `num_classes`, and `epochs`.

```python
#Initialize variables
batch_size = 128
num_classes = 10
epochs = 25
```

```python
model_2 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, kernel_size=(3, 3), padding='same', input_shape=x_train.shape[1:], activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=2),

    tf.keras.layers.Conv2D(64, kernel_size=(3, 3), padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=2),
    
    tf.keras.layers.Conv2D(128, kernel_size=(3, 3), padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=2),
    
    tf.keras.layers.Conv2D(256, kernel_size=(3, 3), padding='same', activation='relu'),
    tf.keras.layers.MaxPooling2D(pool_size=(2, 2),strides=2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(num_classes, activation='softmax'),
])

# Customizable Parameters
model_2.compile(optimizer=tf.keras.optimizers.legacy.RMSprop(learning_rate=0.0005, momentum = 0.9, decay=1e-06),
            loss='categorical_crossentropy', metrics=['acc'])
```
